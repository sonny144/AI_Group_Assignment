```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# Distance matrix
distances = np.array([
    [0, 7, 20, 15, 12],
    [10, 0, 6, 14, 18],
    [20, 6, 0, 15, 30],
    [15, 14, 25, 0, 2],
    [12, 18, 30, 2, 0]
])
```

```python
# coordinates for the places 
coordinates = np.array([
    [4, 3],  # Dorado Park
    [5, 9],  # Khomasdal
    [2, 6],  # Katutura
    [8, 4],  # Eros
    [9, 8]   # Klein Windhoek
])
```

```python
# This Function will calculate total distance
def calculate_total_distance(route):
    return sum(distances[route[i-1], route[i]] for i in range(len(route)))
```

```python
# Function to generate a random initial route
def generate_initial_route(number_of_places):
    route = list(range(number_of_places))
    random.shuffle(route)
    return route
```

```python
# This Function to find a neighbour by swapping two places
def find_neighbour(route):
    neighbour = route.copy()
    a, b = random.sample(range(len(route)), 2)
    neighbour[a], neighbour[b] = neighbour[b], neighbour[a]
    return neighbour
```

```python
# This is the Hill climbing algorithm
def hill_climbing(distances):
    current_route = generate_initial_route(len(distances))
    current_distance = calculate_total_distance(current_route)
```

```python
while True:
        neighbour = find_neighbour(current_route)
        neighbour_distance = calculate_total_distance(neighbour)
        
        if neighbour_distance < current_distance:
            current_route, current_distance = neighbour, neighbour_distance
        else:
            return current_route, current_distance
```

```python
# Hill climbing algorithm
def hill_climbing(distances):
    current_route = generate_initial_route(len(distances))
    current_distance = calculate_total_distance(current_route)
    
    while True:
        neighbour = find_neighbour(current_route)
        neighbour_distance = calculate_total_distance(neighbour)
        
        if neighbour_distance < current_distance:
            current_route, current_distance = neighbour, neighbour_distance
        else:
            # Correctly indented return statement
            return current_route, current_distance
```

```python
# Visualization function
def visualize_route(route, coordinates):
    plt.figure(figsize=(10, 6))
    # Plot the points
    for i in range(len(route)):
        plt.plot(coordinates[route[i-1:i+1], 0], coordinates[route[i-1:i+1], 1], 'ro-')
    for i, txt in enumerate(route):
        plt.annotate(txt, (coordinates[i][0], coordinates[i][1]))
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.title('Route Visualization')
    plt.show()
```

```python
# Runs the hill climbing algorithm
best_route, best_distance = hill_climbing(distances)
```

```python
# Visualize the route
visualize_route(best_route, coordinates)
```
